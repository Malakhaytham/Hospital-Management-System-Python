```python
class Person:

    def __init__(self, name, person_id, contact):

        self.name = name
        self.person_id = person_id
        self.contact = contact


    def display_info(self):

        print("Name:", self.name)
        print("ID:", self.person_id)
        print("Contact:", self.contact)
```

```python
person1 = Person(
    "Malak",
    101,
    "01000000000"
)

person1.display_info()
print(person1.contact)
```

```python
class Doctor(Person):

    def __init__(self, name, person_id, contact, specialization):

        super().__init__(name, person_id, contact)

        self.specialization = specialization
        self.patients = []


    def diagnose(self):

        print(f"Dr. {self.name} is diagnosing the patient.")


    def prescribe_medicine(self, medicine):

        print(f"Dr. {self.name} prescribed {medicine}.")


    def add_patient(self, patient):

        # Prevent adding the same patient twice
        if patient not in self.patients:

            self.patients.append(patient)

            print("Patient added successfully.")


    # ================= NEW METHOD =================
    def create_visit(self,
                     patient,
                     visit_date,
                     diagnosis,
                     prescription,
                     next_appointment):

        visit = Visit(
            visit_date,
            diagnosis,
            prescription,
            next_appointment
        )


        patient.add_visit(visit)

        patient.add_medical_history(diagnosis)
        patient.add_prescription(prescription)
        patient.add_appointment(next_appointment)

        print("Visit added successfully.")


    def view_patients(self):

        print(f"Patients of Dr. {self.name}:")

        if len(self.patients) == 0:

            print("No patients.")

        else:

            for patient in self.patients:

                print("-", patient.name)
```

```python
class Visit:

    def __init__(self, visit_date, diagnosis, prescription, next_appointment):

        self.visit_date = visit_date
        self.diagnosis = diagnosis
        self.prescription = prescription
        self.next_appointment = next_appointment


    def display_visit(self):

        print("Visit Date:", self.visit_date)
        print("Diagnosis:", self.diagnosis)
        print("Prescription:", self.prescription)
        print("Next Appointment:", self.next_appointment)
```

```python
class GeneralDoctor(Doctor):

    def __init__(self, name, person_id, contact):

        super().__init__(
            name,
            person_id,
            contact,
            "General"
        )

    def diagnose(self):
        print(f"Dr. {self.name} is diagnosing common illnesses.")
```

```python
class SpecialistDoctor(Doctor):

    def __init__(self, name, person_id, contact, specialization):

        super().__init__(
            name,
            person_id,
            contact,
            specialization
        )

    def diagnose(self):
        print(
            f"Dr. {self.name} is performing a specialist diagnosis in {self.specialization}."
        )
```

```python
class Patient(Person):

    def __init__(self, name, person_id, contact):

        super().__init__(name, person_id, contact)

        self.medical_history = []
        self.prescriptions = []
        self.appointments = []
        self.visits = []          # Store all patient visits

        self.doctor = None


    def assign_doctor(self, doctor):

        self.doctor = doctor

        doctor.add_patient(self)


    def add_medical_history(self, record):

        self.medical_history.append(record)


    def add_prescription(self, medicine):

        self.prescriptions.append(medicine)


    def add_appointment(self, appointment):

        self.appointments.append(appointment)


    # ================= NEW METHOD =================
    def add_visit(self, visit):

        self.visits.append(visit)


    def view_medical_history(self):

        print("Medical History:")

        for record in self.medical_history:

            print("-", record)


    def view_prescriptions(self):

        print("Prescriptions:")

        for medicine in self.prescriptions:

            print("-", medicine)


    def view_appointments(self):

        print("Appointments:")

        for appointment in self.appointments:

            print("-", appointment)


    # ================= NEW METHOD =================
    def view_visits(self):

        print("Visits:")

        if len(self.visits) == 0:

            print("No visits yet.")

        else:

            for visit in self.visits:

                visit.view_visit()


    def view_patient_info(self):

        print(f"Patient Name : {self.name}")

        print(f"ID : {self.person_id}")

        print(f"Contact : {self.contact}")

        if self.doctor:

            print(f"Doctor : Dr. {self.doctor.name}")

        else:

            print("Doctor : Not Assigned")
```

```python
class Staff(Person):

    def __init__(self, name, person_id, contact, role):

        super().__init__(name, person_id, contact)

        self.role = role


    def schedule_appointment(self):

        print(f"{self.name} scheduled an appointment.")


    def discharge_patient(self):

        print(f"{self.name} discharged the patient.")
```

```python
class Hospital:

    def __init__(self, name):

        self.name = name

        self.doctors = []

        self.patients = []

        self.staff = []


    def add_doctor(self, doctor):

        self.doctors.append(doctor)


    def add_patient(self, patient):

        self.patients.append(patient)


    def add_staff(self, staff):

        self.staff.append(staff)


    def show_doctors(self):

        print("===== Doctors =====")

        for doctor in self.doctors:

            print(
                f"{doctor.name} | {doctor.specialization}"
            )


    def show_patients(self):

        print("===== Patients =====")

        for patient in self.patients:

            if patient.doctor:

                print(
                    f"{patient.name} | Doctor : {patient.doctor.name}"
                )

            else:

                print(
                    f"{patient.name} | Doctor : Not Assigned"
                )


    def show_staff(self):

        print("===== Staff =====")

        for staff in self.staff:

            print(
                f"{staff.name} | {staff.role}"
            )
```

```python
import sqlite3


# Create database connection
connection = sqlite3.connect("hospital.db")

cursor = connection.cursor()


print("Database connected successfully!")
```

```python
# Delete old tables

cursor.execute("DROP TABLE IF EXISTS visits")
cursor.execute("DROP TABLE IF EXISTS patients")
cursor.execute("DROP TABLE IF EXISTS doctors")
cursor.execute("DROP TABLE IF EXISTS staff")
cursor.execute("DROP TABLE IF EXISTS users")


connection.commit()

print("Old tables deleted!")
```

```python
# ================= CREATE TABLES =================


# Doctors Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS doctors(

    doctor_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    contact TEXT,
    specialization TEXT,
    doctor_type TEXT

)
""")


# Patients Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS patients(

    patient_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    contact TEXT,
    age INTEGER,
    gender TEXT,
    doctor_id INTEGER,

    FOREIGN KEY(doctor_id)
    REFERENCES doctors(doctor_id)

)
""")


# Visits Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS visits(

    visit_id INTEGER PRIMARY KEY AUTOINCREMENT,

    patient_id INTEGER,
    doctor_id INTEGER,

    visit_date TEXT,
    diagnosis TEXT,
    prescription TEXT,
    next_appointment TEXT,


    FOREIGN KEY(patient_id)
    REFERENCES patients(patient_id),


    FOREIGN KEY(doctor_id)
    REFERENCES doctors(doctor_id)

)
""")


# Staff Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS staff(

    staff_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    contact TEXT,
    role TEXT

)
""")


# Users Table (Login)
cursor.execute("""
CREATE TABLE IF NOT EXISTS users(

    username TEXT PRIMARY KEY,
    password TEXT,
    role TEXT

)
""")


connection.commit()


print("All tables created successfully!")
```

```python
# ================= INSERT DOCTOR DATA =================

def save_doctor(doctor, doctor_type):

    cursor.execute("""
    INSERT INTO doctors(
        doctor_id,
        name,
        contact,
        specialization,
        doctor_type
    )

    VALUES (?, ?, ?, ?, ?)

    """,
    (
        doctor.person_id,
        doctor.name,
        doctor.contact,
        doctor.specialization,
        doctor_type
    ))


    connection.commit()

    print("Doctor saved successfully!")
```

```python
save_doctor(doctor1, "Normal")

save_doctor(general_doc, "General")

save_doctor(specialist_doc, "Specialist")
```

```python
# ================= INSERT PATIENT DATA =================

def save_patient(patient):

    doctor_id = None

    if patient.doctor:

        doctor_id = patient.doctor.person_id


    cursor.execute("""
    INSERT INTO patients(
        patient_id,
        name,
        contact,
        age,
        gender,
        doctor_id
    )

    VALUES (?, ?, ?, ?, ?, ?)

    """,
    (
        patient.person_id,
        patient.name,
        patient.contact,
        patient.age if hasattr(patient, "age") else None,
        patient.gender if hasattr(patient, "gender") else None,
        doctor_id
    ))


    connection.commit()

    print("Patient saved successfully!")
```

```python
# ================= DOCTOR PATIENTS WINDOW =================

def doctor_patients_window():

    global doctor_search_entry

    doctor_window = tk.Toplevel(root)

    doctor_window.title("Doctor Patients")

    doctor_window.geometry("400x200")

    doctor_window.configure(bg="#EAF4FC")


    tk.Label(

        doctor_window,

        text="Enter Doctor ID",

        font=("Arial", 12, "bold"),

        bg="#EAF4FC"

    ).pack(pady=10)


    doctor_search_entry = tk.Entry(

        doctor_window,

        width=25

    )

    doctor_search_entry.pack(pady=5)


    tk.Button(

        doctor_window,

        text="Search",

        width=20,

        command=show_doctor_patients

    ).pack(pady=20)
```

```python
# ================= SHOW DOCTOR PATIENTS =================

def show_doctor_patients():

    try:

        doctor_id = int(doctor_search_entry.get())

        cursor.execute("""

        SELECT
            name,
            specialization

        FROM doctors

        WHERE doctor_id = ?

        """, (doctor_id,))

        doctor = cursor.fetchone()

        if doctor is None:

            messagebox.showerror(
                "Error",
                "Doctor not found!"
            )
            return

        result = ""
        result += f"Doctor: {doctor[0]}\n"
        result += f"Specialization: {doctor[1]}\n"
        result += "=" * 40 + "\n\n"


        cursor.execute("""

        SELECT
            patient_id,
            name,
            contact

        FROM patients

        WHERE doctor_id = ?

        """, (doctor_id,))

        patients = cursor.fetchall()

        if not patients:

            result += "No patients found."

        else:

            for patient in patients:

                patient_id = patient[0]

                result += f"Patient: {patient[1]}\n"
                result += f"ID: {patient_id}\n"
                result += f"Contact: {patient[2]}\n\n"


                # Medical History
                result += "Medical History:\n"

                cursor.execute("""

                SELECT record

                FROM medical_history

                WHERE patient_id = ?

                """, (patient_id,))

                history = cursor.fetchall()

                if history:

                    for h in history:
                        result += f"• {h[0]}\n"
                else:
                    result += "No History\n"

                result += "\n"


                # Prescriptions
                result += "Prescriptions:\n"

                cursor.execute("""

                SELECT medicine

                FROM prescriptions

                WHERE patient_id = ?

                """, (patient_id,))

                medicines = cursor.fetchall()

                if medicines:

                    for m in medicines:
                        result += f"• {m[0]}\n"
                else:
                    result += "No Prescriptions\n"

                result += "\n"


                # Visits
                result += "Visits:\n"

                cursor.execute("""

                SELECT
                    visit_date,
                    diagnosis,
                    prescription,
                    next_appointment

                FROM visits

                WHERE patient_id = ?

                """, (patient_id,))

                visits = cursor.fetchall()

                if visits:

                    for visit in visits:

                        result += (
                            f"Date: {visit[0]}\n"
                            f"Diagnosis: {visit[1]}\n"
                            f"Prescription: {visit[2]}\n"
                            f"Next Appointment: {visit[3]}\n"
                            "--------------------------\n"
                        )

                else:

                    result += "No Visits\n"

                result += "\n" + "=" * 40 + "\n\n"

        messagebox.showinfo(
            "Doctor Patients",
            result
        )

    except Exception as e:

        messagebox.showerror(
            "Error",
            str(e)
        )
```

```python
# ================= SAVE MEDICAL HISTORY =================

def save_medical_history(patient_id, record):

    cursor.execute("""

    INSERT INTO medical_history(

        patient_id,
        record

    )

    VALUES (?, ?)

    """,

    (

        patient_id,
        record

    ))

    connection.commit()


# ================= SAVE PRESCRIPTION =================

def save_prescription(patient_id, medicine):

    cursor.execute("""

    INSERT INTO prescriptions(

        patient_id,
        medicine

    )

    VALUES (?, ?)

    """,

    (

        patient_id,
        medicine

    ))

    connection.commit()
```

```python
# ================= SAVE VISIT =================

def save_visit(patient, visit):

    cursor.execute("""

    INSERT INTO visits(

        patient_id,
        doctor_id,
        visit_date,
        diagnosis,
        prescription,
        next_appointment

    )

    VALUES (?, ?, ?, ?, ?, ?)

    """,

    (

        patient.person_id,
        patient.doctor.person_id,
        visit.visit_date,
        visit.diagnosis,
        visit.prescription,
        visit.next_appointment

    ))

    connection.commit()


    # Save Medical History

    save_medical_history(

        patient.person_id,

        visit.diagnosis

    )


    # Save Prescription

    save_prescription(

        patient.person_id,

        visit.prescription

    )


    print("Visit saved successfully!")
```
